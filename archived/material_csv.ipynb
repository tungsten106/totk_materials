```python
import pandas as pd

# 创建表格的函数
def create_table(data):
    rows = [row.split('：') for row in data.split('\n') if '：' in row]
    columns = rows[0]
    rows = rows[1:]
    df = pd.DataFrame(rows, columns=columns)
    return df

# 将内容按套装分割
sets = [set.strip() for set in content.split('「') if set.strip()]

# 创建每个套装的表格
tables = []
for set in sets:
    table = create_table(set)
    tables.append(table)

# 打印每个表格
for i, table in enumerate(tables, start=1):
    print(f'套装{i}的升级材料表格:')
    print(table)
    print('\n')
```
